```python
%matplotlib inline

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import random
import math
```

```python
uniform_data = stats.uniform.rvs(size=100000,  # Generate 100000 numbers
                                 loc = 0,      # From 0
                                 scale=10)
```

```python
pd.DataFrame(uniform_data).plot(kind="density",  # Plot the distribution
                               figsize=(11,11),
                               xlim=(-5,15));
```

```python
cumilative=stats.uniform.cdf(x=2.5,         # Cutoff value (quantile) to check
                  loc=0,         # Distribution start
                  scale=10)
print(cumilative)
```

```python
for x in range(-1,12):
  print("pdf at "+str(x))
  print(stats.uniform.pdf(x,loc=0,scale=10))
```

```python
cdf=[]
for x in range(-1,12):
  print("cdf at "+str(x))
  print( stats.uniform.cdf(x, loc=0, scale=10) )
  cdf.append(stats.uniform.cdf(x,loc=0,scale=10))
```
